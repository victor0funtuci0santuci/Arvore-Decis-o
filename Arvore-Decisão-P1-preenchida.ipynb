````python
# 1 Criando um dataframe com os nomes de cada variavel.
metadados = pd.DataFrame({'Variável': data.columns, 'Tipo de Dados': data.dtypes})
metadados
````

````python
# 2 Adicione uma coluna nesse dataframe chamada "qtd_categorias" 
#e coloque nela o número de categorias correspondente de cada variável.

# Criando a coluna "qtd_categorias".
qtd_categorias = []

for variavel in data.columns:
    qtd_categorias.append(data[variavel].nunique())

metadados['qtd_categorias'] = qtd_categorias
metadados
````
